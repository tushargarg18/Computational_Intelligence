# Example 3

Draw a graf of the function $y=f(x)=x^2$ and its derivatice $y'=f'(x). Use Tensoflow for derivative calculation.

```python
# Example
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

@tf.function
def f(x):
  y = x*x
  return y

@tf.function
def dfdx(x):
  y = f(x)
  return tf.gradients(y, x)[0]

x = tf.Variable([-3.,-2.,-1.,0.,1.,2.,3.])
print(f(x).numpy())
print(dfdx(x).numpy())

plt.plot(f(x))
plt.plot(dfdx(x))
```

# Task 2

Draw a graf of the function $y=\sigma(x)=\frac{1}{1+e^{-x}}$ and its derivative $y'=s'(x)$. Use Tensoflow for derivative calculation.

```python
import tensorflow as tf
import matplotlib.pyplot as plt

@tf.function
def graf(x):
  return 1 / ( 1 + tf.exp(-x))

@tf.function
def graf_dx(x):
  return tf.gradients(graf(x), x)[0]


X = tf.linspace(0, 16, 1000)

y = graf(X)
dy = graf_dx(X)

X = X.numpy()
y = y.numpy()
dy = dy.numpy()

plt.figure(figsize = (12,6))
plt.plot(X, y, color = 'Blue')
plt.plot(X,dy, color = 'Red')
plt.title('graf function plot')
plt.grid(True)
plt.show()

```

# Task 4

Create a simple neural feed forward neural network using Tensorflow.

Input layer -> hidden layer -> output layer

for IRIS dataset

```python
Write your code here
```

```python
import pandas as pd
import numpy as np

from google.colab import drive

drive.mount('/content/drive')


data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/iris_processed.csv')

print(data.head())
```

```python
x_features = data.iloc[:, 1:5].to_numpy()
y_labels = data.iloc[:, 5:].to_numpy()

print(x_features.shape)
print(y_labels.shape)
```

```python
# normalize

x_min = x_features.min()
x_max = x_features.max()

x_features = (x_features - x_min)/ (x_max - x_min)

print(x_features[:5])
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x_features,
    y_labels,
    test_size = 0.2,
    random_state = 42
)

print(X_train.shape)
print(X_test.shape)
```

```python
model = models.Sequential([
    layers.Input(shape = (4,)),
    layers.Dense(8, activation = 'relu'),
    layers.Dense(3, activation = 'softmax')
])
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs = 150, batch_size = 8, verbose = 1)
```

```python
loss, acc = model.evaluate(X_test, y_test)

print(loss)
print(acc)
```
